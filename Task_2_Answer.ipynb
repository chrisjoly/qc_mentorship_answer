```python
q = QuantumRegister(2, 'q')
c = ClassicalRegister(2, 'c')
circ = QuantumCircuit(q, c)
```

```python
pi = 3.14159265359



circ.rx(pi/2, q[0])

circ.cx(q[0],q[1])

circ.measure(q,c)
```

```python
circ.draw()
```

```python
shots = [1024,4096,8192,65536]

# results = []
# res = []
i = len(shots)
leg = []


jobs = []
r = []
count = []

for x in range(0, i):    
    jobs.append(execute(circ, backend=Aer.get_backend('qasm_simulator'), shots=shots[x]))
    r.append(jobs[x].result())
    count.append(r[x].get_counts(circ))
    print(count[x])
    leg.append(str(shots[x])+" shots")
    
    
```
